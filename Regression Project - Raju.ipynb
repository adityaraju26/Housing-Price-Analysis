```python
modified_df = df.dropna(axis=1)
```

```python
modified_df
```

```python
X = modified_df[['LotArea', 'EnclosedPorch']]
y = modified_df.SalePrice
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
print('Coefficients: ', lm.coef_)
print('Intercept: ', lm.intercept_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
line_prediction = lm.predict(X_test)
```

```python
print('Coefficients: ', lm.coef_)
print('Mean squared error: %.2f'
    %mean_squared_error(y_test, line_prediction))
print('R2 Score: %.2f'
    %mean_squared_error(y_test, line_prediction, squared=False))
print('Coefficient of determination: %.2f'
     %r2_score(y_test, line_prediction))
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("train.csv")
```

```python
X = df[['OverallQual', 'OverallCond']]
y = df.SalePrice
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
print('Coefficients: ', lm.coef_)
print('Intercept: ', lm.intercept_)
```

```python
line_prediction = lm.predict(X_test)
```

```python
print('Coefficients: ', lm.coef_)
print('Mean squared error: %.2f'
    %mean_squared_error(y_test, line_prediction))
print('R2 Score: %.2f'
    %mean_squared_error(y_test, line_prediction, squared=False))
print('Coefficient of determination: %.2f'
     %r2_score(y_test, line_prediction))
```
